#### Unique games in the stream table

```python
%%sql

SELECT DISTINCT game
FROM stream;
```

#### Unique channels in the stream table

```python
%%sql

SELECT DISTINCT channel
FROM stream;
```

#### The most popular games in the stream table

```python
%%sql

SELECT game, COUNT(*) as number_of_viewers
FROM stream
GROUP BY game
ORDER BY number_of_viewers DESC;
```

#### List of countries and their number of LoL viewers

```python
%%sql

SELECT country, COUNT(*) as number_of_viewers
FROM stream
WHERE game = 'League of Legends'
GROUP BY country
ORDER BY number_of_viewers DESC;
```

####  List of players and their number of streamers

```python
%%sql

SELECT player, COUNT(*) as number_of_streamers
FROM stream
GROUP BY player
ORDER BY number_of_streamers DESC;
```

#### Creating a new column genre for each of the games

```python
%%sql

SELECT game,
CASE
    WHEN game = 'League of Legends' THEN 'MOBA'
    WHEN game = 'Dota 2' THEN 'MOBA'
    WHEN game = 'Heroes of the Storm' THEN 'MOBA'
    WHEN game = 'Counter-Strike: Global Offensive' THEN 'FPS'
    WHEN game = 'DayZ' THEN 'Survival'
    WHEN game = 'ARK: Survival Evolved' THEN 'Survival'
    ELSE 'Other'
END as genre,
COUNT(*) as number_of_viewers
FROM stream
GROUP BY game
ORDER BY number_of_viewers DESC;
```

#### View count change in the course of a day

```python
%%sql

SELECT strftime('%H', time) as hour, COUNT(*) as count
FROM stream
WHERE country = 'PL'
GROUP BY hour;
```

#### Joining stream and chat

```python
%%sql

SELECT *
FROM stream
JOIN chat
    ON stream.device_id = chat.device_id
LIMIT 10;
```

# Data visualization

```python
import matplotlib.pyplot as plt
```

```python
game_viewers = stream.groupby('game').time.count().reset_index()
game_viewers.rename(columns={'time': 'number_of_viewers'}, inplace=True)
game_viewers = game_viewers.sort_values(by='number_of_viewers', ascending=False)
game_viewers
```

```python
plt.figure(figsize=(12,10))
plt.bar(range(len(game_viewers.head(10).game)), game_viewers.head(10).number_of_viewers, color='slateblue')
plt.title('Featured Games Viewers')
plt.legend(['Twitch'])
plt.xlabel('Games')
plt.ylabel('Viewers')
ax = plt.subplot()
ax.set_xticks(range(10))
ax.set_xticklabels(game_viewers.head(10).game, rotation=90)
plt.show()
plt.clf()
```

```python
countries = stream[stream['game']=='League of Legends'].groupby('country').time.count().reset_index()
countries.rename(columns={'time': 'number_of_viewers'}, inplace=True)
countries = countries.sort_values(by='number_of_viewers', ascending=False)
countries.head(8)

labels = ['US', 'CA', 'DE', 'GB', 'TR', 'AU', 'SE', 'NL', 'Others']
```

```python
top = countries[:8].copy()
others = pd.DataFrame(data = {
    'country' : ['others'],
    'number_of_viewers' : [countries['number_of_viewers'][8:].sum()]
})
countries = pd.concat([top, others])
```

```python
plt.figure(figsize=(8,10))
plt.pie(countries.number_of_viewers, autopct='%1.0f%%', pctdistance=1.1)
plt.title("League of Legends Viewers' Whereabouts")
plt.legend(labels, loc="right")
plt.axis('equal')
plt.show()
plt.clf()
```

```python
stream.time = pd.to_datetime(stream.time)
stream['hour'] = stream.time.dt.hour
```

```python
hours = stream[stream['country'] == 'PL'].groupby('hour').login.count().reset_index()
hours.rename(columns={'login': 'number_of_viewers'}, inplace=True)
```

```python
plt.figure(figsize=(12,10))
plt.plot(hours.hour, hours.number_of_viewers)
plt.title('Time Series in Poland')
plt.xlabel('Hour')
plt.ylabel('Viewers')
ax = plt.subplot()
ax.set_xticks(hours.hour)
ax.set_yticks([0, 200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800, 2000])
y_upper = [i + (i*0.15) for i in hours.number_of_viewers]
y_lower = [i - (i*0.15) for i in hours.number_of_viewers]
 
plt.fill_between(hours.hour, y_lower, y_upper, alpha=0.2)
plt.show()
plt.clf()
```
